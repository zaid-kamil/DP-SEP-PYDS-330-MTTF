```python
import numpy as np
import matplotlib.pyplot as plt
```

`plt.show()` should only be written at the end of your graph code. It is not necessary to write it after every graph. You can also use `plt.show()` to display multiple graphs at once.

#### a simple graph with sine wave and cosine wave

```python
t = np.linspace(0, 10, 100)
fig = plt.figure(figsize=(20,6))
plt.plot(t, np.sin(t), 'r-', label='sine wave of t')
plt.plot(t, np.cos(t), 'b--', label='cos(t)')
plt.legend()
plt.show()
```

A Figure object is the outermost container for a Matplotlib plot. The Figure object contain multiple Axes objects. So, the Figure is the final graphic that may contain one or more Axes. The Axes represent an individual plot.

```python
fig, ax = plt.subplots(nrows= 1, ncols= 2, figsize=(20,6))
ax[0].plot(t, np.sin(t), 'r-', label='sine wave of t')
ax[1].plot(t, np.cos(t), 'b--', label='cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows= 2, ncols= 1, figsize=(20,6))
ax[0].plot(t, np.sin(t), 'r-', label='sine wave of t')
ax[1].plot(t, np.cos(t), 'b--', label='cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show()
```

#### adjusting the labels, ticks, legend and x-axis, y-axis

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Sine wave')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
plt.xlabel('x values', fontsize=14, fontdict={'color': 'green'})
plt.ylabel('y values',fontsize=14, fontdict={'color': 'green'})
plt.title('Sine wave',fontsize=16, fontdict={'color': 'green', 'weight': 'bold',})
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
plt.xlabel('x values', fontsize=14, fontdict={'color': 'green'})
plt.ylabel('y values',fontsize=14, fontdict={'color': 'green'})
plt.title('Sine wave',fontsize=16, fontdict={'color': 'green', 'weight': 'bold',})
plt.text(4, .5, 'y = sin(x)', fontsize=10, fontdict={'color': 'red',})
plt.text(1, .25, 'crest', fontsize=10, fontdict={'color': 'red',})
plt.text(4, -.5, 'trough', fontsize=10, fontdict={'color': 'red',})
# annotate
plt.annotate('local max', xy=(1.6, 1), xytext=(1, 1.5), arrowprops={'facecolor': 'green'})
plt.annotate('local min', xy=(4.6, -.99), xytext=(7, -.75), arrowprops={'facecolor': 'green'})
plt.show()
```

```python
# changing the background color
fig = plt.figure(figsize=(20,6), facecolor='lightblue')
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'b', label='sine wave of t') 
# change axes background color
ax = plt.gca() # get the current graph 'axes'
ax.set_facecolor('#8943ff')
plt.show()
```

```python
# scatter plot
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, marker='o', color='red')
plt.show()
```

```python
# bubble plot
fig = plt.figure(figsize=(20,6))
x = np.random.randn(100)
y = np.random.randn(100)
size = np.random.randint(10,1000,100)
plt.scatter(x, y, marker='*', color='red', s=size, alpha=.5,edgecolors='black')
plt.show()
```

```python
# bubble plot
fig = plt.figure(figsize=(20,6))
x = np.random.randn(100)
y = np.random.randn(100)
size = np.random.randint(10,1000,100)
plt.scatter(x, y, marker='o', s=size, c=size, alpha=.5,edgecolors='black', cmap='gist_rainbow')
plt.colorbar()
plt.show()
```

```python
gender_count = np.array([120,45])
plt.pie(gender_count)
plt.show()
```

```python
gender_count = np.array([120,45])
gender_labels = np.array(['Female','Male'])
props={'color': 'white', 'weight': 'bold', 'fontsize': 14}
plt.pie(gender_count, labels= gender_labels, textprops=props) # text prop
plt.show()
```

```python
plt.pie(gender_count, labels= gender_labels, textprops=props, autopct='%.1f%%') # autopct 
plt.show()
```

```python
continents = ['Africa', 'Asia', 'Europe', 'North America', 'South America', 'Oceania']
population = [1216, 4436, 738, 579, 422, 38]
props={'color': 'white', 'weight': 'bold', 'fontsize': 8}
plt.pie(population, labels=continents, autopct='%.1f%%', textprops=props)
plt.show()
```

```python
# DONUT CHART
fig = plt.figure(figsize=(6,6), facecolor=None)
continents = ['Africa', 'Asia', 'Europe', 'North America', 'South America', 'Oceania']
population = [1216, 4436, 738, 579, 422, 38]
tprops={'color': 'yellow', 'weight': 'bold', 'fontsize': 8}
wprops={'width': 0.6, 'linewidth': 2}
plt.pie(population, labels=continents, autopct='%.1f%%', textprops=tprops, wedgeprops=wprops)
plt.show()
```

# visualization
- *Univariate analysis*
    - univarite analysis is the simplest form of analyzing data. In univariate analysis, only one variable(single column) is taken into consideration at a time. It is used to find patterns in single variable data.
- *Bivariate analysis*
    - Bivariate analysis is the simultaneous analysis of two variables. It is used to find the relationship between two variables.
- *Multivariate analysis*
    - Multivariate analysis is the simultaneous analysis of three or more variables. It is used to find the relationship between three or more variables.

```python
# bar plot
sales = np.array([100, 200, 300, 400, 500, 420, 300, 500, 600])
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep'])
plt.bar(x=months, height=sales)
plt.show()
```

```python
# bar chart with text
plt.bar(x=months, height=sales)
for i in range(len(sales)):
    plt.text(x=i, y=10, s=sales[i], ha='center', fontdict={'color':'white'})
plt.show()
```

```python
plt.barh(y= months, width=sales)
plt.show()
```

```python
plt.bar(x=months, height=sales, color='green', edgecolor='black', alpha=.5)
```

```python
colors = ['red', 'green', 'blue', 'yellow',
          'orange', 'purple', 'brown', 'pink', 'gray']
plt.bar(x=months, height=sales, color=colors, edgecolor='black', alpha=.5)
```

```python
plt.bar(x=months,
    height=sales, 
    color=colors, 
    edgecolor='black', 
    alpha=.5,
    hatch='/')
```

```python
# box plot
fig = plt.figure(figsize=(20,6))
x = np.random.randint(1,10000, (100, 3))
plt.boxplot(x, meanline=True, showmeans=True, showfliers=True, notch=True)
plt.show()
```

```python
# box plot
s1_marks = [95,83,94,95,72]
s2_marks = [45,93,90,65,72]
s3_marks = [75,63,70,55,62]
marks = np.array([s1_marks, s2_marks, s3_marks]).T
marks
```

```python
plt.boxplot(marks, labels=['S1', 'S2', 'S3'], showmeans=True, meanline=True)
plt.show()
```

```python
# stem plot
fig = plt.figure(figsize=(20,6))
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.stem(x, y, linefmt='r:', markerfmt='r*', basefmt='g')
plt.show()
```

```python
# hexbin plot
fig = plt.figure(figsize=(8,6))
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=16, cmap='YlOrRd')
plt.colorbar()
plt.show()
```
