#### dataframe reading and selecting subsets of data
- row wise selection
    - `df.loc[]`
    - `df.iloc[]`
    - `df.head()`
    - `df.tail()`
- column wise selection
    - `df['column_name']`
    - `df[['column_name1', 'column_name2']]`
    - `df.query()`

```python
df.loc[100] # specific row, as a series
```

```python
df.loc[100:120] # multiple rows, as a dataframe
```

```python
df.loc[[51, 101, 201, 301]] # multiple rows in a list style
```

```python
cities = [51, 101, 201, 301, 401]
cols = ['City','Population']
df.loc[cities, cols] # multiple rows and columns specified
```

```python
df.loc[[51, 101, 201, 301, 401], ['City','Population']] # same as above
```
